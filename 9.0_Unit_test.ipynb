```python

importlib.reload(socialmediasite)


class TestSiteBasic(unittest.TestCase):
    attr=None

    def setUp(self):
        print(f"Setting up {self.__class__.__name__}")
        self.site = socialmediasite.SocialMediaSite(b,browser='chrome')
        self.browser = self.site.browser
        doctest.DocTestSuite()
        # self.customer_2 = Customer('Tina', 'Smith', 3000)
        
    def test_herokuapp(self):
        _url="https://testpages.herokuapp.com/styled/index.html"
        self.browser.goto(_url)
        self.assertEqual(self.browser.url,_url)
        self.assertEqual(self.browser.h1().text,"Test Pages For Automating")
        self.assertEqual(self.browser.link(id="fileuploadtest").text,"File Upload Example Page")
        
    
    def tearDown(self):
        print("tearDown")
        self.browser.close()
        pass

    @unittest.skipIf(os.name=='nt',"posix script")
    def test_nonWindows(self):
        self.assertEqual(os.name,'posix')

def test_all():
    suite = unittest.TestSuite()
    for m in ["utils","socialmediasite"]:
        suite.addTests(doctest.DocTestSuite(m))
    
    suite.addTests(unittest.makeSuite(TestSiteBasic))

    # for example in glob.glob(os.path.join(os.getcwd(), "examples", "*", "*.yml")):
    #     suite.addTest(GilesCompilationTestCase(example))

    return suite


unittest.TextTestRunner(verbosity=2
                       # ).run(
                       ).debug(
    test_all())
```
