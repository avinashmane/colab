```python
from requests.structures import CaseInsensitiveDict
def APIcall(path,params):
    token=SPApiProxy._PySendPulse__token
    # url = 'https://api.sendpulse.com/whatsapp/account'
    url = 'https://api.sendpulse.com'+path
    headers = CaseInsensitiveDict()
    headers["Authorization"] = f"Bearer {token}"
    # payload={}#
    print(headers)
    try:
        resp = requests.post(url, headers=headers,data=params)                    
    except Exception as e:
        print(f"Error {e!r}")
    return  resp.json()

APIcall('/whatsapp/contacts/send',{"contact_id":avinash,"message": 
    {
      "type": "text",
      "text": {
        "body": "string"
      }
    }
  })
```
